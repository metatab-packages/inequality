# Housing

Among people who own a home, Blacks have less home equity at every income category, and the amount of eqity does not increase much with income for the bottom 70th percentile. 


```python
t = dfs[(dfs.housecl == 1) & (dfs.nincpctle <= 90)]

with new_plot('Mean Home Equity of Homeowners By Income Group and Race', source,
             ylabel='Home Equity ($)', xlabel='Normal Income Percentile') as (fig,ax):

    t.groupby(['nincpctle', 'race']).homeeq.mean().unstack().astype(int).plot(ax=ax)
```

```python
t = dfs[ (dfs.nincpctle <= 90)]

t['nwmhousing'] = t.networth - t.homeeq

with new_plot('Networth and Home Equity By Income Group and Race', source,ylabel='',
              xlabel='Normal Income Percentile', panes=3) as (fig,ax):

    t.groupby(['nincpctle', 'race']).homeeq.median().unstack().astype(int).plot(ax=ax[0])
    ax[0].set_title('Median Home Equity', fontsize=16)
    ax[0].set_ylabel('Home Equity ($)')
    
    t.groupby(['nincpctle', 'race']).networth.median().unstack().astype(int).plot(ax=ax[1])
    ax[1].set_title('Median Net Worth', fontsize=16)
    ax[1].set_ylabel('Net worth ($)')
    
    t.groupby(['nincpctle', 'race']).nwmhousing.median().unstack().astype(int).plot(ax=ax[2])
    ax[2].set_title('Median Networth Minus Home Equity', fontsize=16)
    ax[2].set_ylabel('$')
```

Removing housing does not reduce the ratios of white to black networth, although it does reduce the amount of the difference. 

```python
def f(df):
    return df.white / df.black

f(t.groupby([ 'race']).networth.median()), f(t.groupby([ 'race']).nwmhousing.median())
```

```python
def f(df):
    return (df.white - df.black)

f(t.groupby([ 'race']).networth.median()), f(t.groupby([ 'race']).nwmhousing.median())
```
